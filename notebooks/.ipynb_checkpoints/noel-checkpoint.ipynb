```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from wordcloud import ImageColorGenerator
from wordcloud import STOPWORDS
```

```python
#Load dataset, /Users/noelfeisst/Downloads/Recipe_dataset.csv
data = pd.read_csv('~/code/bkpearse/Cooking-Project/raw_data/recipe_dataset.csv')
```

```python
data.head()
```

```python
#rename or should I delete first column
data.rename(columns= {'Unnamed: 0' : 'index'}, inplace=True)
```

```python
#Check for duplicates
data.duplicated().sum()
```

```python
#check for missing data
data.isnull().sum()
```

```python
#creating wordcloud for title
text = ' '.join(i for i in data.title)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color='white').generate(text)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()

```

```python
#creating wordcloud for ingredients
text = ' '.join(i for i in data.ingredients)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color='white').generate(text)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
